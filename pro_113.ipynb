```python
#importing all libraries

import seaborn as sns
import pandas as pd
import csv
import statistics
import plotly.figure_factory as ff
import random 

#uploding file to colab

from google.colab import files 
data_to_load = files.upload()

#reading file and creating database

df = pd.read_csv("savings_data_final.csv")
sns.boxplot(data=df,x=df["quant_saved"])

#code for iqr

q1 = df['quant_saved'].quantile(0.25)
q3 = df['quant_saved'].quantile(0.75)
iqr = q3-q1
print(f"Q1 - {q1} Q3 = {q3} IQR = {iqr}")

#code for lower whisker and upper whisker

lower_whisker = q1 - 1.5*iqr  
upper_whisker = q3 + 1.5*iqr 
print(f"LW = {lower_whisker} UW = {upper_whisker}")

#creating new database

new_df = df[df['quant_saved']<upper_whisker]
all_savings = new_df['quant_saved'].tolist()

#finding mean, median ,mode ,std_dev

mean = statistics.mean(all_savings)
std_deviation = statistics.stdev(all_savings)


print(f"mean of savings - {mean}")
print(f"std of savings - {std_deviation}")
print(f"median of savings - {statistics.median(all_savings)}")
print(f"mode of savings - {statistics.mode(all_savings)}")

first_std_deviation_start,first_std_deviation_end = mean - std_deviation,mean+std_deviation
print(f"First (start)- {first_std_deviation_start}, First (end)- {first_std_deviation_end},")

second_std_deviation_start,second_std_deviation_end = mean -(2*std_deviation) , mean + ( 2*std_deviation)
print(f"Second (start)- {second_std_deviation_start}, Second (end)- {second_std_deviation_end},")

third_std_deviation_start,third_std_deviation_end = mean -(3*std_deviation) , mean + ( 3*std_deviation)
print(f"Third (start)- {third_std_deviation_start}, Third (end)- {third_std_deviation_end},")

#creating graph
fig = ff.create_distplot([new_df['quant_saved'].tolist()], ['Savings'] , show_hist = False )
fig.show()
```
